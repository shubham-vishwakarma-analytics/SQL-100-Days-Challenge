## Overfitting vs Underfitting

Now comes one of the most important ML concepts


## 🔴 Overfitting
* Model learns TOO much (memorizes data)
* → Works well on training
* → Fails on new data

## 🔵 Underfitting
* Model learns TOO little
* → Performs badly everywhere

## 1. High train accuracy + low test accuracy
### 👉 Overfitting ✔️

* ✔️ Model memorizes training data
* ❌ Fails on new/unseen data

## 2. Poor performance on both
### 👉 Underfitting ✔️

* ❌ Model didn’t learn enough
* ❌ Performs badly everywhere

## Real-Life Example

🔴 Overfitting

* You memorize answers of 100 questions
* Exam comes → different questions → you fail
  
🔵 Underfitting

* You didn’t study at all
* Exam comes → you fail anyway

 ## Key Difference 
* Overfitting → Too complex model  
* Underfitting → Too simple model

## 🔴 How to Reduce Overfitting

👉 At least 2 strong methods:

### 1. Use More Data
More data → model learns general patterns → less memorization
### 2. Simplify the Model
Reduce complexity (e.g., lower degree, less depth)
### 3. Train-Test Split / Cross Validation
Helps detect overfitting early
### 4. Regularization (L1 / L2)
Penalizes complex models

## 🔵 How to Fix Underfitting
### 1. Increase Model Complexity
Use more powerful model (e.g., Linear → Polynomial, Tree depth ↑)
### 2. Add More Features
More useful inputs → better learning
### 3. Train Longer (if applicable)
Model needs more learning time

## 🧠 Simple Memory Trick
### Overfitting → Too complex → Simplify / More data  
### Underfitting → Too simple → Make model stronger
